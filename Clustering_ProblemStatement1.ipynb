Perform Clustering(Hierarchical, Kmeans & DBSCAN) for the crime data and identify the number of clusters formed and draw inferences.

Data Description:
Murder- Muder rates in different places of United States
Assualt- Assualt rate in different places of United States
UrbanPop- urban population in different places of United States
Rape- Rape rate in different places of United States


```python
import pandas as pd 
import matplotlib.pyplot as plt 
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN 
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")
```

### Hierarchical Clustering

```python
crime_data=pd.read_csv('crime_data.csv')
crime_data
```

```python
crime_data.info()
```

```python
crime_data.isnull().sum()
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(crime_data.iloc[:,1:])
df_norm
```

```python
#Creating Dendogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='complete'))
```

```python
#Creating Clusters
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'complete')
hc
```

```python
y_hc = hc.fit_predict(df_norm)
y_hc
```

```python
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
crime_data['Clusters']=Clusters
crime_data
```

```python
plt.figure(figsize=(10,7))  
plt.scatter(crime_data['Clusters'],crime_data['Murder'], c=y_hc)
```

```python
crime_data[crime_data['Clusters']==0]
```

```python
crime_data[crime_data['Clusters']==1]
```

```python
crime_data[crime_data['Clusters']==2]
```

### Kmeans Clustering

```python
crime1_data=pd.read_csv('crime_data.csv')
crime1_data
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(crime1_data.iloc[:,1:])
df_norm
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
```

```python
model.labels_
```

```python
#Assign clusters to the data set
crime1_data['clusterid_new'] = model.labels_
crime1_data
```

```python
model.cluster_centers_
```

```python
crime1_data.groupby('clusterid_new').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10,7))  
plt.scatter(crime1_data['clusterid_new'],crime1_data['Murder'], c=model.labels_)
```

```python
crime1_data[crime1_data['clusterid_new']==1]
```

```python
crime1_data[crime1_data['clusterid_new']==2]
```

### DBSCAN Clustering

```python
crime1=pd.read_csv('crime_data.csv')
crime1
```

```python
scaler = StandardScaler()
scaled_crime_data = scaler.fit_transform(crime1.iloc[:,1:])
scaled_crime_data
```

```python
dbs=DBSCAN(min_samples=5,eps=2)
```

```python
dbs.fit(scaled_crime_data)
```

```python
dbs.labels_
```

```python
Cluster=pd.DataFrame(dbs.labels_,columns=['Clust'])
Cluster
```

```python
crime1['Cluster']=Cluster
crime1
```

```python
plt.figure(figsize=(10,7))  
plt.scatter(crime1['Cluster'],crime1['Murder'], c=dbs.labels_)
```
